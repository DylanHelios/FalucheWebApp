# Tableau 2 Colonnes

```python
from docx import Document

# Charger le document Word
doc = Document("Code-National-136-mis-au-propre.docx")

# Ouvrir un fichier texte en Ã©criture
with open("resultat.txt", "w", encoding="utf-8") as f:
    # Parcourir tous les tableaux du document
    for i, table in enumerate(doc.tables):
        for row in table.rows:
            # RÃ©cupÃ©ration du texte de chaque cellule et suppression des sauts de ligne
            row_data = [cell.text.replace("\n", " ").strip() for cell in row.cells]
            print(row_data)

            # VÃ©rification : sâ€™assurer quâ€™il y a au moins 4 colonnes
            if len(row_data) < 2:
                continue

            # On suppose que la premiÃ¨re cellule contient une liste sÃ©parÃ©e par virgules ou points-virgules
            chaine = row_data[0]
            tableau = [element.strip() for element in chaine.replace(";", ",").split(",") if element.strip()]

            # Pour chaque Ã©lÃ©ment du tableau, Ã©crire dans le fichier
            for element in tableau:
                # On enregistre les 3 premiÃ¨res colonnes utiles
                ligne = f"{element};{row_data[1]};{row_data[2]}\n"
                f.write(ligne)

print("âœ… DonnÃ©es enregistrÃ©es dans 'resultat.txt'")
```

# Convertisseur JPEG /JPG to PNG

```python
%pip install pillow
import os
from PIL import Image

def convertir_jpeg_en_png(dossier_entree, dossier_sortie=None):
    """
    Convertit toutes les images .jpg/.jpeg d'un dossier en .png.

    Args:
        dossier_entree (str): chemin du dossier contenant les images source.
        dossier_sortie (str, optionnel): dossier oÃ¹ enregistrer les PNG.
                                         Si None, les PNG sont enregistrÃ©s dans le mÃªme dossier.

    Exemple :
        convertir_jpeg_en_png("images/")
        convertir_jpeg_en_png("images/source", "images/png")
    """
    if not os.path.isdir(dossier_entree):
        print(f"âŒ Dossier introuvable : {dossier_entree}")
        return

    if dossier_sortie is None:
        dossier_sortie = dossier_entree
    else:
        os.makedirs(dossier_sortie, exist_ok=True)

    nb_converties = 0
    for fichier in os.listdir(dossier_entree):
        chemin_fichier = os.path.join(dossier_entree, fichier)
        if not os.path.isfile(chemin_fichier):
            continue

        nom, ext = os.path.splitext(fichier)
        ext = ext.lower()

        if ext in ['.jpg', '.jpeg']:
            try:
                with Image.open(chemin_fichier) as img:
                    img = img.convert("RGBA")  # prÃ©serve transparence potentielle
                    chemin_sortie = os.path.join(dossier_sortie, f"{nom}.png")
                    img.save(chemin_sortie, "PNG")
                    nb_converties += 1
                    print(f"âœ… Converti : {fichier} â†’ {nom}.png")
            except Exception as e:
                print(f"âš ï¸ Erreur conversion {fichier} : {e}")

    print(f"\nâœ… Conversion terminÃ©e : {nb_converties} image(s) convertie(s).")

dossier_entree = "xxx"
convertir_jpeg_en_png(dossier_entree)

```

```python
from PIL import Image
import os

def concat_images_from_folder(folder_path, output_path, image_names=None):
    # Si image_names n'est pas fourni, prendre les 3 premiers fichiers du dossier
    if image_names is None:
        files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        files.sort()
        image_names = files[:3]
    images = [Image.open(os.path.join(folder_path, name)) for name in image_names]

    # Redimensionner toutes les images Ã  la mÃªme hauteur (hauteur de la plus petite)
    min_height = min(img.height for img in images)
    resized_images = [img.resize((int(img.width * min_height / img.height), min_height)) for img in images]

    total_width = sum(img.width for img in resized_images)
    result = Image.new('RGBA', (total_width, min_height))

    x_offset = 0
    for img in resized_images:
        result.paste(img, (x_offset, 0))
        x_offset += img.width

    result.save(output_path)
    print(f"âœ… Image composÃ©e sauvegardÃ©e : {output_path}")

# Exemple d'utilisation
concat_images_from_folder("zzz/Bac SVT", "zzz/Bac SVT/Bac SVT.png")
```

```python
import fitz  # PyMuPDF
import os

folder_name = "image_faluche"


def extract_images_with_names(pdf_path, page_number, output_folder="image_faluche"):
    
    prev_image_name = None
    index = 0
    
    create_folder(output_folder)

    # Ouvrir le PDF
    doc = fitz.open(pdf_path)
    if page_number <= len(doc):
        page = doc.load_page(page_number - 1)  # Les pages sont indexÃ©es Ã  partir de 0

        # Obtenir les images avec leurs rÃ©fÃ©rences
        images = page.get_images(full=True)
        text_instances = page.get_text("dict")  # Obtenir le texte et ses positions

        # Extraire les informations de texte et leurs positions
        text_blocks = [block for block in text_instances["blocks"] if block["type"] == 0]  # Type 0 signifie texte

        # Pour chaque image, trouver le texte Ã  sa gauche
        for img_index, img in enumerate(images):
            xref = img[0]  # La rÃ©fÃ©rence de l'image

            try:
                base_image = doc.extract_image(xref)
                image_bytes = base_image["image"]
                image_ext = base_image["ext"]

                # Obtenir les coordonnÃ©es de l'image
                image_rects = page.get_image_rects(xref)
                if not image_rects:
                    continue

                image_rect = image_rects[0]  # Supposons qu'il y a une seule occurrence de l'image

                # Trouver le texte le plus proche Ã  gauche de l'image
                closest_text = None
                min_distance = float('inf')

                for block in text_blocks:
                    for line in block["lines"]:
                        for span in line["spans"]:
                            text = span["text"].strip()
                            text_rect = span["bbox"]  # CoordonnÃ©es du texte (x0, y0, x1, y1)

                            # VÃ©rifier si le texte est Ã  gauche de l'image et sur la mÃªme ligne verticale
                            if text_rect[2] < image_rect.x0:  # Le texte se termine avant le dÃ©but de l'image
                                vertical_center_text = (text_rect[1] + text_rect[3]) / 2
                                vertical_center_image = (image_rect.y0 + image_rect.y1) / 2
                                vertical_distance = abs(vertical_center_text - vertical_center_image)

                                # Si les centres verticaux sont proches, considÃ©rer que le texte est sur la mÃªme ligne
                                if vertical_distance < 20:  # Seuil pour la distance verticale
                                    distance = abs(text_rect[2] - image_rect.x0)
                                    if distance < min_distance:
                                        min_distance = distance
                                        closest_text = text

                # Utiliser le nom correspondant
                if closest_text:

                    image_name = closest_text

                    image_name = image_name.replace("Â«", "").replace("Â»", "").replace(" ", "")
                    image_name = image_name.replace("Lettres", "")


                    if closest_text:
                        image_name = closest_text.strip()
                        image_name = image_name.replace("Â«", "").replace("Â»", "").replace(" ", "").replace(":", "")
                        image_name = image_name.replace("Lettres", "")

                        # ğŸ” Regarder le suivant
                        next_text = None
                        if img_index + 1 < len(images):
                            # On pourrait aussi extraire son closest_text, mais pour simplifier :
                            next_text = image_name  # <-- Ã  adapter selon ta logique d'extraction du texte

                        # VÃ©rifier si le prochain est identique
                        if next_text == image_name:
                            image_name = f"{image_name}"
                            index += 1
                        else:
                            index = 0  # reset si le suivant est diffÃ©rent
                    else:
                        image_name = f"image_{img_index}"

                # Sauvegarder l'image
                with open(os.path.join(output_folder, f"{image_name}.{image_ext}"), "wb") as image_file:
                    image_file.write(image_bytes)

                print(f"Image sauvegardÃ©e sous {os.path.join(output_folder, f'{image_name}.{image_ext}')}")

            except ValueError as e:
                print(f"Erreur avec l'image {img_index}: {e}")
                continue

        return images
    else:
        print(f"Le fichier PDF n'a pas {page_number} pages.")
        return None

# Exemple d'utilisation
pdf_path = PATH_CODE
page_number = 15
images = extract_images_with_names(pdf_path, page_number)
```

```python
from PIL import Image
import os


def replace_black_background_two_colors(image_path, output_path, top_color="#9900cc", bottom_color="#ff0000", threshold=30):
    """
    Remplace le fond noir d'une image par deux couleurs avec une coupure nette horizontale :
    - moitiÃ© supÃ©rieure -> top_color
    - moitiÃ© infÃ©rieure -> bottom_color
    """
    img = Image.open(image_path).convert("RGBA")
    width, height = img.size
    new_data = []

    # Convertir hex -> RGB
    top_rgb = Image.new("RGB", (1,1), top_color).getpixel((0,0))
    bottom_rgb = Image.new("RGB", (1,1), bottom_color).getpixel((0,0))

    for y in range(height):
        for x in range(width):
            r, g, b, a = img.getpixel((x, y))
            if r < threshold and g < threshold and b < threshold:
                # Choisir couleur selon moitiÃ©
                new_data.append((*top_rgb, 255) if y < height // 2 else (*bottom_rgb, 255))
            else:
                new_data.append((r, g, b, a))

    img.putdata(new_data)

    # Forcer PNG
    if not output_path.lower().endswith(".png"):
        output_path = os.path.splitext(output_path)[0] + ".png"

    img.save(output_path)
    print(f"âœ… Image sauvegardÃ©e : {output_path}")

# Exemple d'utilisation
replace_black_background_two_colors(
    "image_faluche/",
    Satin + "Sciences politiques",
    top_color=BLEU,
    bottom_color=NOIR,
    threshold=30
)
```
